Code to get the image urls for sets, parts, and minifigs using the rebrickable api

```
import pandas as pd
pd.set_option('display.max_colwidth', None)

data = pd.read_csv("/content/drive/My Drive/550_project/minifigs.csv")
data.head()
```

```
import requests
import time

headers = {"Authorization": "key 2d80b7f6b0e2a2ecc64d23f73bca5483"}


next = "https://rebrickable.com/api/v3/lego/minifigs/?page=1&page_size=1000"

while next:
  r = requests.get(next, headers=headers)
  response = r.json()
  next = response["next"]
  minifigs = response["results"]
  for minifig in minifigs:
    data.loc[data["fig_num"] == minifig["set_num"], "image_url"] = minifig["set_img_url"]
  time.sleep(10)

data.to_csv("/content/drive/My Drive/550_project/minifigs.csv")
```

```
import requests
import time

headers = {"Authorization": "key 2d80b7f6b0e2a2ecc64d23f73bca5483"}


next = "https://rebrickable.com/api/v3/lego/parts/?page=1&page_size=1000"

part_num = []
name = []
image_url = []

while next:
  r = requests.get(next, headers=headers)
  response = r.json()
  next = response["next"]
  parts = response["results"]
  for part in parts:
    part_num.append(part["part_num"])
    name.append(part["name"])
    image_url.append(part["part_img_url"])
  time.sleep(10)

print(len(part_num))
print(len(name))
print(len(image_url))

data = {"part_num": part_num, "name": name, "image_url": image_url}

parts = pd.DataFrame(data)

parts.to_csv("/content/drive/My Drive/550_project/parts.csv")
```

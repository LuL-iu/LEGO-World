```
import pandas as pd
import re
pd.set_option('display.max_colwidth', None)


reviews = pd.read_csv("/content/drive/Shared drives/CIS 550 Group Project/User Reviews Data/final clean reviews.csv")

def remove_html_tags(text):
    return re.sub("<.*?>", "", text)

#Remove html tags
reviews["review"] = reviews["review"].map(remove_html_tags)

#remove not needed columns
reviews.drop(["Name"], axis=1, inplace=True)

reviews.to_csv("/content/drive/Shared drives/CIS 550 Group Project/User Reviews Data/final clean reviews.csv", index=False)

reviews.head()
```
